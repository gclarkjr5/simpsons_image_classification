# Import Libraries

```python
import scipy as sp
import tensorflow as tf
from tensorflow import keras
import os
import numpy as np
import matplotlib.pyplot as plt
import plotly.plotly as plotly
```

# Get Train Set

```python
main_directory = '/media/sf_Gee_Hard_Drive/Kaggle Data/the-simpsons-characters-dataset/simpsons_dataset/simpsons_dataset'
classes = os.listdir(main_directory)[0:2]
class_indices = np.arange(len(classes))

class_dict = dict(zip(classes, class_indices))
class_dict
```

```python
picObj = []
for dirs in classes:
    directory = main_directory + '/' + dirs + '/'
    pics = os.listdir(directory)
    for pic in pics:
        path = directory + pic
        pix = sp.misc.imread(path)
        picObj.append([pix, dirs, class_dict[dirs]])
        
picObj = np.array(picObj)
picObj.shape, picObj[0][0].shape
```

# Print out the first 10 pics

```python
%matplotlib inline

plt.figure(figsize=(10,10))
for i in range(len(picObj[0:10])):
    plt.subplot(5,2,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid('off')
    plt.imshow(picObj[i][0], cmap=plt.cm.binary)
    plt.xlabel(picObj[i][1])
```

# Filter for same dimensions

```python
dim = (432, 288, 3)
picFilt = list(filter(lambda x: x[0].shape == dim, picObj))
picFilt = np.array(picFilt)
picFilt.shape
```

# Verify images are same dims

```python
%matplotlib inline

plt.figure(figsize=(10,10))
for i in range(len(picFilt[0:10])):
    plt.subplot(5,2,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid('off')
    plt.imshow(picFilt[i][0], cmap=plt.cm.binary)
    plt.xlabel(picFilt[i][1])
```

# PreProcess and convert 0-255 RGB to 0-1 range

```python
for i in range(len(picFilt)):
    picFilt[i][0] = picFilt[i][0]/255
    
picFilt.shape, picFilt[0][2], picFilt[0][1], picFilt[0][0]
```

```python
train_images = list(map(lambda x: x[0], picFilt))
train_images = np.array(train_images)

train_labels = list(map(lambda x: x[2], picFilt))
train_labels = np.array(train_labels)

train_images.shape, train_labels.shape
```

# Model Construction

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(432, 288, 3)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
model.compile(optimizer=tf.train.AdamOptimizer(), 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs=5)
```
